```python
import math
import pandas as pd
import numpy as np
from pprint import pprint
from tqdm import tqdm
```

```python
races = pd.read_csv('races.csv')
results = pd.read_csv('results.csv')
qualifying = pd.read_csv('qualifying_results.csv')
driver_standings = pd.read_csv('driver_standings.csv')
constructor_standings = pd.read_csv('constructor_standings.csv')
```

```python
results.head()
```

```python
df1 = pd.merge(races, results, how='inner', on=['season', 'round', 'circuit_id', 'url']).drop(['points', 'status'], axis = 1)
df1.head()
```

```python
df2 = pd.merge(df1, driver_standings, how='left', on=['season', 'round', 'driver']) 
df2.head()
```

```python
df3 = pd.merge(df2, constructor_standings, how='left', on=['season', 'round', 'constructor']) #from 1958
df3.head()
```

```python
qualifying.rename(columns = {'pos': 'grid'}, inplace = True)
qualifying.drop(['driver'], axis = 1, inplace = True)
final_df = pd.merge(df3, qualifying, how='inner', on=['season', 'round', 'grid']).drop(['car'], axis = 1)
```

```python
final_df[:-20]
```

```python
from dateutil.relativedelta import *

final_df['date'] = pd.to_datetime(final_df.date)
final_df['date_of_birth'] = pd.to_datetime(final_df.date_of_birth)
final_df['driver_age'] = final_df.apply(lambda x: relativedelta(x['date'], x['date_of_birth']).years, axis=1)
final_df.drop(['date', 'date_of_birth'], axis = 1, inplace = True)

final_df.head()
```

```python
for col in ['driver_points', 'driver_wins', 'driver_standings_pos', 'constructor_points', 
            'constructor_wins' , 'constructor_standings_pos']:
    final_df[col] = final_df[col].fillna(0)
```

```python
final_df.isnull().sum().sum()
```

```python
final_df['qualifying_time'] = final_df.qualifying_time.map(lambda x: 120 if str(x) == '00.000' else(float(str(x).split(':')[1]) + (60 * float(str(x).split(':')[0])) if x != 0 else 0))
final_df =pd.get_dummies(final_df, columns = ['circuit_id', 'nationality', 'constructor'] )

# remove outliers
for col in final_df.columns:
    if 'nationality' in col and final_df[col].sum() < 70:
        final_df.drop(col, axis = 1, inplace = True)
    elif 'constructor' in col and final_df[col].sum() < 70:
        final_df.drop(col, axis = 1, inplace = True)
    elif 'circuit_id' in col and final_df[col].sum() < 35:
        final_df.drop(col, axis = 1, inplace = True)
```

```python
final_df.sort_values(['season', 'round', 'grid'], inplace = True)
```
